# Part 1

## Problem 1: n-Queen Problem

```python
# Custom fitness function
def queens_max(state):    
    fitness = 0
    for i in range(len(state) - 1):
        for j in range(i + 1, len(state)):
            if (state[j] != state[i]) and (state[j] != state[i] + (j - i)) and (state[j] != state[i] - (j - i)):
                fitness += 1

    return fitness
```

```python
# Define a fitness function
fitness_cust = mlrose.CustomFitness(queens_max)
```

```python
# Define an Optimization Problem
problem_low = mlrose.DiscreteOpt(length = 8, fitness_fn = fitness_cust, maximize = True, max_val = 8)
problem_high = mlrose.DiscreteOpt(length = 40, fitness_fn = fitness_cust, maximize = True, max_val = 40)
```

```python
init_state_low = np.arange(8)
init_state_high = np.arange(40)
```

### Randomized Hill Climbing

```python
# Low complexity
start_time = time.time()
_, best_fitness, fitness_curve = mlrose.random_hill_climb(problem_low, max_attempts=10, max_iters=100, restarts=0, init_state=init_state_low, curve=True, random_state=None)
print(best_fitness)
print("--- %s seconds ---" % (time.time() - start_time))

plt.plot(fitness_curve)
plt.ylabel('Fitness')
plt.xlabel('Iterations')
plt.title('Randomized Hill Climbing (Small Problem Size)')
```

```python
# Low complexity
start_time = time.time()
_, best_fitness, fitness_curve = mlrose.random_hill_climb(problem_high, max_attempts=10, max_iters=100, restarts=0, init_state=init_state_high, curve=True, random_state=None)
print(best_fitness)
print("--- %s seconds ---" % (time.time() - start_time))

plt.plot(fitness_curve)
plt.ylabel('Fitness')
plt.xlabel('Iterations')
plt.title('Randomized Hill Climbing (Large Problem Size)')
```

```python
# Read cancer data
cancer_data = pd.read_csv("cancer_data.csv", header=0)
print ("Cancer data read successfully!")
cancer_data.head()
```

```python
n_cases = cancer_data.shape[0]

n_features = len(list(cancer_data.columns[:-1]))

n_malignant = cancer_data[cancer_data.diagnosis == 'M'].shape[0]

n_benign = cancer_data[cancer_data.diagnosis == 'B'].shape[0]

malignancy_rate = (float(n_malignant)/n_benign) * 100

# Print the results
print ("Total number of cases: {}".format(n_cases))
print ("Number of features: {}".format(n_features))
print ("Number of cases which were positive: {}".format(n_malignant))
print ("Number of students which were negative: {}".format(n_benign))
print ("Malignancy rate in patients: {:.2f}%".format(malignancy_rate))
```

```python
cancer_data.drop('id',axis=1,inplace=True)
cancer_data.drop('Unnamed: 32',axis=1,inplace=True)
cancer_data['diagnosis'] = cancer_data['diagnosis'].map({'M':1,'B':0})
cancer_data.describe()
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(cancer_data[cancer_data.columns[1:11]], cancer_data[cancer_data.columns[0]], test_size = 0.2, random_state = 3)

# Normalize feature data
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
def plot_confusion_matrix(cm, classes,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

## Gradient Descent

```python
# Initialize neural network object and fit object
nn_model1 = mlrose.NeuralNetwork(hidden_nodes = [3], activation = 'relu', \
                                 algorithm = 'gradient_descent', max_iters = 1000, \
                                 bias = True, is_classifier = True, learning_rate = 0.0001, \
                                 early_stopping = True, clip_max = 5, max_attempts = 100, \
                                 random_state = 3)

start_time = time.time()
nn_model1.fit(X_train_scaled, y_train)
print("Training Time \n--- %s seconds ---" % (time.time() - start_time))
```

```python
plt.plot(nn_model1.fitted_weights)
plt.title('Gradient Descent')
```

```python
# Predict labels for train set and assess accuracy
y_train_pred = nn_model1.predict(X_train_scaled)

y_train_accuracy = accuracy_score(y_train, y_train_pred)

print(y_train_accuracy)

# Predict labels for test set and assess accuracy
y_test_pred = nn_model1.predict(X_test_scaled)

y_test_accuracy = accuracy_score(y_test, y_test_pred)

print(y_test_accuracy)
```

```python
plt.plot(nn_model1.predicted_probs)
plt.title('Gradient Descent')
plt.xlabel('Instances')
plt.ylabel('Confidence (Probability)')
```

```python
confusion = confusion_matrix(y_test, y_test_pred)

TP = confusion[1, 1]
TN = confusion[0, 0]
FP = confusion[0, 1]
FN = confusion[1, 0]

print ("True Positive: ", TP)
print ("True Negative: ", TN)
print ("False Positive: ", FP)
print ("False Negative: ", FN)

plt.figure()
plot_confusion_matrix(confusion, classes=[0,1], title='Gradient Descent')
```
